**Описание технологического процесса**

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

**1. Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов. На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

**Описание данных:**

*Технологический процесс*

- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    - Xanthate — ксантогенат (промотер, или активатор флотации);
    - Sulphate — сульфат (на данном производстве сульфид натрия);
    - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

*Параметры этапов*

- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи
- Наименование признаков
- Наименование признаков должно быть такое: [этап].[тип_параметра].[название_параметра] Пример: rougher.input.feed_ag

*Возможные значения для блока [этап]:*

- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики
Возможные значения для блока [тип_параметра]:

- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка расчета эффективности обогащения на обучающей выборке для признака rougher.output.recovery. Вычисление MAE между расчитанным признаком и значением имеющегося признака.

```python
#рассчитаем recovery самостоятельно
train['recovery'] = (train['rougher_output_concentrate_au']*(train['rougher_input_feed_au'] -train['rougher_output_tail_au']))/(train['rougher_input_feed_au']*(train['rougher_output_concentrate_au']-train['rougher_output_tail_au']))*100


#  рассчитаем МАЕ
mae=mean_absolute_error(train['recovery'],train['rougher_output_recovery'])
print ('MAE составляет:\n',mae)
```

Вывод:

МАЕ в районе 0 говорит о том, что расчёт эффективности проведён в датафрейме верно.

### Функция для вычисления итоговой sMAPE.

```python
# Создадим функцию для расчёта SMAPE.

def smape(target, predictions):
  '''Функция принимает прогнозы и целевые признаки для расчёта метрики качества. '''
  smape = abs(target - predictions) / ((abs(target) + abs(predictions)) / 2) *100
  smape.fillna(0, inplace=True)
  smape=smape.mean()
  return smape
```

```python
# Создадим итоговую функцию расчета SMAPE для 2х целевых показателей.

def final_smape(smape_rougher,smape_final) :
  ''' Функция рассчитывает итоговую метрику
симметричного среднего абсолютного процентного отклонения
по двум показателям SMAPE целевых признаков.
  '''
  final_smape_coef = .25*smape_rougher+.75*smape_final
  return round (final_smape_coef, 2)
```

```python
# Подготовим выборки для обучения моделей.

features_col = test.columns.tolist()[1:-2]
target_col = test.columns.tolist()[:-3:-1]
f_train, t_train=new_train [features_col], new_train [target_col]
f_test, t_test=test[features_col], test[target_col]
```

### Обучение различных моделей и их оценка кросс-валидацией. Выбер лучшей модели и проверка её на тестовой выборке.

```python
# Функция для расчёта средней метрики SMAPE для модели.

def cv_score (model, target_name='rougher', features=f_train,target=t_train):
  '''Функция для оценки качества модели с применением SMAPE. '''
  target_choice=target['rougher_output_recovery'] if target_name=='rougher' else target['final_output_recovery']
  cvs = cross_val_score(model,features, target_choice, scoring= smape_coef,n_jobs=-1)
  return  round (abs(cvs. mean()),2)
```

```python
# Создадим функцию для подбора гиперпараметров и обучения на них модели.

def random_search(model,parameters,target_name='rougher',features=f_train, target=t_train):
  target_choice=target['rougher_output_recovery'] if target_name=='rougher' else target['final_output_recovery']
  rs= RandomizedSearchCV(estimator = model, param_distributions = parameters,cv = 3, n_iter=5,n_jobs = -1, verbose = 2,scoring=smape_coef)
  rs.fit(features,target_choice)
  rs_param=rs.best_params_
  rs_score=round(abs(rs.best_score_),2)
  return rs_param, rs_score
```
