### Знакомство с данными. Переименование колонок

```python
# создадим функцию для переименования колонок.
def new_name_for_col (col : list) -> list:
  ''' Функция принимает в себя список названий колонок датафрейма и
возвращает список с преобразованными названиями
(все буквы в нижнем регистре, названия из нескольких слов соединены "_")
  '''
  glue = '_'
  new_col=[]
  for i in col:
    new_col. append(''.join(glue + x.lower() if x.isupper() else x.lower() for x in i)[1:])
  return new_col
```

### Обучение моделей Случайного леса, Дерева решений без учёта дисбаланса классов.

```python
# Создадим функцию для расчёта основных коэффициентов.

def coeff(model,target=t_valid,features=f_valid):
  ''' Функция для расчёта и вывода
коэффициентов: F1, recall, precision,confusion_matrix.

  '''
  if target is t_valid:
    print ('F1 (Valid) : ',round (f1_score(target, model. predict (features))
, 4))
    print ('Recall (Valid): ',round (recall_score(target, model. predict (features))
, 4))
    print ('Precision (Valid): ',round (precision_score(target, model. predict (features))
, 4))
    print ()
    print('Confusion_matrix\n',confusion_matrix(target, model. predict (features)))
  else:
    print ('F1 (Test) : ',round (f1_score(target, model. predict (features))
, 4))
    print ('Recall (Test): ',round (recall_score(target, model. predict (features))
, 4))
    print ('Precision (Test): ',round (precision_score(target, model. predict (features))
, 4))
    print ()
    print('Confusion_matrix\n',confusion_matrix(target, model. predict (features)))
```

```python
def mod_coef(model,target=t_valid,features=f_valid):
  ''' Функция для расчёта
коэффициентов F1, recall, precision.

  '''
  f1=round(f1_score(target, model. predict (features)), 4)
  rec=round(recall_score(target, model. predict (features)), 4)
  prec=round(precision_score(target, model. predict (features)), 4)
  return f1, rec, prec
```

```python
# Создадим функцию для построения ROC-кривой.

def plot_roc_curve(fpr, tpr):
    '''Функция для построения ROC-кривой
    '''
    plt.plot(fpr, tpr, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.legend()
    plt.show()
```

```python
#  Создадим функцию для построения одновременного построения графика ROC  и расчёта roc_auc_score.

def roc_auc_and_plot(target,probabilities_one_valid):
  '''Функция стргит ROC-кривую и выводит результат расчёта коэффициента roc_auc_score.
  '''
  fpr, tpr, thresholds = roc_curve(target,probabilities_one_valid)
  plot_roc_curve(fpr, tpr)
  auc_roc= round (roc_auc_score(target,probabilities_one_valid), 4)
  print('AUC-ROC :',auc_roc)
```

```python
#  Создадим функцию расчета roc_auc_score.

def roc_auc(model,target=t_valid,features=f_valid):
  '''Функция возвращает результат расчёта коэффициента roc_auc_score.
  '''
  auc_roc= round (roc_auc_score(target,model.predict_proba(features)[:, 1]), 4)
  return auc_roc
```

### Использование техники 'upsample' и 'downsample' для уравновешивания классов.

    Преобразование в технике "upsample" проходит в несколько этапов:

    - Разделить обучающую выборку на отрицательные и положительные объекты;
    - Скопировать несколько раз положительные объекты;
    - С учётом полученных данных создать новую обучающую выборку;
    - Перемешать данные: идущие друг за другом одинаковые вопросы не помогут обучению.

```python
# Создадим функцию для балансировки классов методом upsample.

def upsample(features, target, repeat):
  '''Функция возвращает перемешанный датафрейм признаков и
серию целевого атрибута, где увеличены данные миноритарного
 класса для достижения баланса в классах
  '''

  features_zeros = features[target == 0]
  features_ones = features[target == 1]
  target_zeros = target[target == 0]
  target_ones = target[target == 1]

  features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
  target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)

  features_upsampled, target_upsampled = shuffle(features_upsampled,
target_upsampled,
random_state=12345)

  return features_upsampled, target_upsampled

# проверим результат применения функции к выборкам для обучения.
f_upsampled, t_upsampled = upsample(f_train, t_train, 4)
t_upsampled.value_counts(normalize=1)
```

```python
# Создадим функцию для балансировки классов методом downsample.

def downsample(features, target, fraction):

    '''Функция возвращает перемешанный датафрейм признаков и
серию целевого атрибута, где уменьшены данные мажоритарного
 класса для достижения баланса в классах
    '''
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)

    return features_downsampled, target_downsampled
# Проверим результат применения функции.
f_downsampled, t_downsampled=downsample(f_train, t_train, 0.25)
t_downsampled.value_counts(normalize=1)
```
